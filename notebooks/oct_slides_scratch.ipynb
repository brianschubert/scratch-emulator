# LHS vs Grid


```python
import matplotlib.pyplot as plt
import numpy as np

n = 6
a = np.linspace(0, 1, 6)

mesh_x, mesh_y, mesh_z = np.meshgrid(a, a, a, indexing="ij")

# 2d

fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(mesh_x, mesh_y, 180)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_title("2D", weight="bold", fontsize=38)
# ax.set_axis_off()

# 3d
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={"projection": "3d"})
ax.scatter(mesh_x, mesh_y, mesh_z, s=120, alpha=0.9)
# ax.set_xlabel("Input 1")
# ax.set_ylabel("Input 2")
# ax.set_zlabel("Input 3")
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
# ax.set_axis_off()
ax.set_title("3D", weight="bold", fontsize=38)
```
